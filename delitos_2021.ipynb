Cambiar nombres de columnas

```python
df.rename(columns={"uso_armas":"uso de armas"})
```

```python
df.columns
```

```python
df = df.dropna(subset=['comuna'])
```

```python
df.comuna = df.comuna.astype(int)
```

```python
df.dtypes
```

```python
df.dia.unique()
```

```python
df.dia = df.dia.replace({"domingo":"Domingo","lunes":"Lunes","viernes":"Viernes","jueves":"Jueves", "martes":"Martes","sábado":"Sábado","miércoles":"Miércoles"})
```

```python
df.dia.value_counts().plot(kind="bar")
```

Tipos de delitos

```python
df.tipo.unique()
```

```python
df2 = df.groupby(['mes'])['tipo'].count().reset_index(name='cantidad')
df2
```

```python
import plotly.io as pio
pio.templates
```

```python
fig = px.bar(df2, x='mes', y='cantidad',
             color="cantidad",
             template='gridon',
             labels={'cantidad':'Cantidad de casos', 'mes':"Mes"},
             title='Cantidad de delitos por mes',
             height=400
             )
fig.show()
```

```python
df3 = df.groupby(['dia'])['tipo'].count().reset_index(name='cantidad')
df3
```

```python
fig2 = px.bar(df3, x='dia', y='cantidad',
             color="cantidad",
             template='plotly',
             labels={'cantidad':'Cantidad de casos', 'dia':"Día"},
             title='Cantidad de delitos por día',
             height=400
             )
fig2.show()
```

```python
df4 = df[(df.tipo == 'Homicidio')]
df4.head()
```

```python
df5 = df4.groupby(['barrio'])['tipo'].count().reset_index(name='Homicidio').sort_values(by='Homicidio', ascending=False)
df5
```

```python
df6 = df.groupby(['barrio'])['tipo'].count().reset_index()
df7 = df6[df6.tipo >= 200]
df7.head()
```

```python
df8 = df.groupby(['comuna'])['tipo'].count().reset_index().sort_values(by='tipo', ascending=False)
df8.head()
```

```python
df8.tail()
```

Cantidad de casos por delito

```python
df9 = df['tipo'].value_counts(ascending=False).reset_index()
df9.head()
```

```python
df9.columns = ['Tipo de delito', 'Cantidad']
df9
```

```python
fig3 = px.bar(df9, x='Tipo de delito', y='Cantidad',
             color="Cantidad",
             template='xgridoff',
             labels={'cantidad':'Cantidad de casos', 'dia':"Día"},
             title='Cantidad de delitos por día',
             height=400
             )
fig3.show()
```
